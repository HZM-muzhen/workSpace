# 自动小说生成工作流演示

本演示展示如何使用 NovelGenerator 自动生成小说章节，每次生成新章节时会回顾之前的内容。

```python
# 1. 加载环境变量
from dotenv import load_dotenv
load_dotenv()
```

```python
# 2. 导入小说生成器
from novel_generator import NovelGenerator
```

## 第一步：初始化小说生成器

创建一个 NovelGenerator 实例，需要提供：
- 小说标题
- 小说简介/背景设定
- 人物设定

```python
# 3. 创建小说生成器
novel = NovelGenerator(
    novel_title="星辰彼岸",
    description="一部科幻冒险小说，讲述人类探索宇宙深处的故事。主角陈宇是一名年轻的宇航员，在一次意外中穿越到了未知的星系，开始了一段惊心动魄的星际冒险。",
    characters="""
陈宇：男主角，28岁，宇航员，勇敢坚毅，对宇宙充满好奇。
林薇：女主角，26岁，首席科学家，聪明冷静，负责飞船的技术支持。
阿尔法：AI助手，先进的人工智能系统，陪伴主角完成冒险。
神秘文明：宇宙中古老的外星文明，掌握着超越人类想象的科技。
"""
)

# 查看小说信息
print(novel.get_novel_summary())
```

## 第二步：生成第一章

调用 `generate_next_chapter()` 方法生成新章节。第一次调用会生成第一章。

```python
# 4. 生成第一章
chapter1 = novel.generate_next_chapter()

# 查看章节内容
print(f"\n第{chapter1['number']}章 {chapter1['title']}")
print("=" * 50)
print(chapter1['content'])
```

## 第三步：生成第二章

再次调用 `generate_next_chapter()` 生成第二章。系统会自动回顾第一章的内容，确保故事的连贯性。

```python
# 5. 生成第二章
chapter2 = novel.generate_next_chapter()

# 查看章节内容
print(f"\n第{chapter2['number']}章 {chapter2['title']}")
print("=" * 50)
print(chapter2['content'])
```

## 第四步：查看章节列表

可以随时查看已生成的所有章节。

```python
# 6. 列出所有章节
novel.list_chapters()
```

## 第五步：继续生成更多章节

重复调用 `generate_next_chapter()` 可以继续生成后续章节。每次生成时，系统都会回顾最近3章的内容。

```python
# 7. 生成第三章
chapter3 = novel.generate_next_chapter()

# 查看章节内容
print(f"\n第{chapter3['number']}章 {chapter3['title']}")
print("=" * 50)
print(chapter3['content'])
```

## 工作流总结

### 自动小说生成工作流的核心流程：

1. **初始化**：创建 NovelGenerator 实例，提供小说的基本设定（标题、简介、人物）

2. **生成章节**：调用 `generate_next_chapter()` 方法
   - 系统自动计算下一章的章节号
   - 回顾之前的章节内容（最近3章）
   - 调用 DeepSeek 模型生成章节标题
   - 调用 DeepSeek 模型生成章节内容
   - 自动保存章节到文件

3. **文件管理**：
   - 每个章节保存为独立的 `.txt` 文件
   - 文件命名格式：`章节号_章节标题.txt`
   - 所有章节保存在 `novels/小说标题/` 目录下

4. **上下文保持**：
   - 每次生成新章节时，都会将之前的章节内容作为上下文传入
   - 确保故事的连贯性和一致性
   - 章节较多时，只传递最近3章的摘要，避免上下文过长

### 使用建议：

1. **先完善设定**：在开始生成前，尽可能详细地描述小说背景和人物设定
2. **逐步生成**：每次生成一章后，可以阅读并调整下一章的生成策略
3. **手动干预**：如果对生成的内容不满意，可以手动修改章节文件，下次生成会基于修改后的内容继续
4. **定期保存**：系统会自动保存章节文件，但建议定期备份整个 novels 目录
