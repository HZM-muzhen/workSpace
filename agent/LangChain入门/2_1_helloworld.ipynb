```python
# 1.加载环境变量
from dotenv import load_dotenv
load_dotenv()
```

```python
from langchain.tools import tool

# 2.定义工具，基础版，通过注释描述工具
@tool
def getWeather(location: str) -> str:
    """
    Get the weather in a given location.
    Args:
        location: city name or coordinates
    """
    return f"Current weather in {location} is sunny"
```

```python
from langchain_openai import ChatOpenAI
# 3.定义Agent
from langchain.agents import create_openai_tools_agent  # 用这个
from langchain.agents import AgentExecutor
from langchain.prompts import ChatPromptTemplate

prompt = ChatPromptTemplate.from_messages([
    ("system", "你是一个专业的助手，可以使用工具来回答用户问题。"),
    ("user", "{input}"),
    ("placeholder", "{agent_scratchpad}"),
])

# 2. 初始化 DeepSeek 模型（这里必须是对象，不是字符串！）
llm = ChatOpenAI(
    model="deepseek-chat",
    api_key="sk-536a550657da49afb1b2ae2da112c240",  # 把这里改成你的key
    base_url="https://api.deepseek.com/v1",
    temperature=0,
)

agent = create_openai_tools_agent(
    llm, # 模型名称（必须是LangChain支持的模型）
    tools=[getWeather], # 工具集
    prompt=prompt  # 必须传提示词，新版强制要求
)
# 执行器（必须加）
agent_executor = AgentExecutor(agent=agent, tools=[getWeather], verbose=True)
```

```python
# 4.调用模型
print("🚀 正在调用大模型...")
response = agent_executor.invoke({
    "input":  "长沙有什么地方好玩的?"
})
# 5.打印结果
print(response)
```
